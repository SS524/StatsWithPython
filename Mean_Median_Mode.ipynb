```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=[4,5,3,6,8,2,7]
```

```python
np.mean(data)  # Finding mean
```

```python
data2=data.copy() 
```

```python
data2
```

```python
data2.append(61) # adding outlier
```

```python
np.mean(data2)
```

```python
plt.figure(figsize=(12,6))
data_mean=np.mean(data)
data2_mean=np.mean(data2)
sns.kdeplot(data=data,
            color='crimson', label='Before adding outlier', fill=True)
sns.kdeplot(data=data2,
            color='limegreen', label='After adding outlier', fill=True)
plt.axvline(data_mean,c='red')
plt.axvline(data2_mean,c='green')
plt.legend()
plt.title('Nature of Mean value after adding the outlier',fontsize=15)
plt.show()
```

```python
np.median(data) # finding median
```

```python
np.median(data2)
```

```python
plt.figure(figsize=(12,6))
data_median=np.median(data)
data2_median=np.median(data2)
sns.kdeplot(data=data,
            color='crimson', label='Before adding outlier', fill=True)
sns.kdeplot(data=data2,
            color='limegreen', label='After adding outlier', fill=True)
plt.axvline(data_median,c='red')
plt.axvline(data2_median,c='green')
plt.legend()
plt.title('Nature of Median value after adding the outlier',fontsize=15)
plt.show()
```
